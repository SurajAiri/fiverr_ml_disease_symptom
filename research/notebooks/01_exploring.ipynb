```python
symptoms = []
def uniqueSymptoms(symp):
    for s in symp.split(','):
        s = s.strip()
        s = s.lower()
        if s != '' and s not in symptoms:
            symptoms.append(s)
    return symp
```

```python
df['Symptoms'].apply(uniqueSymptoms)
```
