```python
train_path = "../../data/training.csv"
test_path = "../../data/testing.csv"
```

```python
train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)
```

```python
df = pd.concat([train_df,test_df])
df.shape, train_df.shape, test_df.shape
```

```python
pd.options.display.max_seq_items = 2000
```

```python
df.drop(columns=['disease'],axis=0).columns
```

```python
X = df.drop(columns=['disease'],axis=0).values
y = df['disease'].values
len(X), len
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.15,random_state = 32)
```

```python
len(y_train), len(y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
# rf.predict(X_test)
```

```python
def evaluate_model(model,name=''):
    print('==='*20)
    print(name)
    # print('==='*20)
    y_pred = model.predict(X_test)
    print('Accuracy: ',accuracy_score(y_test,y_pred))
    print(confusion_matrix(y_test,y_pred))
    print(classification_report(y_test,y_pred))
```

```python
rf = RandomForestClassifier()
svc = SVC()
knn = KNeighborsClassifier(n_neighbors = 5)
dtc = DecisionTreeClassifier()
```

```python
rf.fit(X_train,y_train)
svc.fit(X_train,y_train)
knn.fit(X_train,y_train)
dtc.fit(X_train,y_train)
```

```python
evaluate_model(rf,"Random Forest")
evaluate_model(svc, "SVC")
evaluate_model(knn,"KNN")
evaluate_model(dtc,"Decision Tree")
```
